###**Condition Checking**

```python
a = 55
b = 55
if a>b:
    print(str(a) + " > " + str(b))
elif a == b:
    print(str(a) + " == " + str(b))
else:
    print(str(a) + " < " +str(b))
```

```python
idx = 0
while(idx < len(arr)):
    print(arr[idx])
    idx = idx + 1
```
